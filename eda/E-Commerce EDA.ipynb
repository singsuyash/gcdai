# Product table cleaning

```python
#description length and name length doesn't make sense, possibly masked data
products.drop("product_description_lenght", axis=1, inplace=True)
products.drop("product_name_lenght", axis=1, inplace=True)
```

```python
#product_category_name, product_description_lenght, product_name_lenght, product_photos_qty; all have missing values for exactly 610 same rows
cat_name_na = products["product_category_name"].isna()
photo_qty_na = products["product_photos_qty"].isna()
na_index = products[(cat_name_na) & (photo_qty_na)].index
print(na_index) # outputs 610
previous_rows = products.shape[0]
products.drop(index=na_index, axis=0, inplace=True)
print("Previous rows: {}, Current rows: {}, Diff: {}".format(str(previous_rows), str(products.shape[0]), str(previous_rows - products.shape[0])))
```

* Replace brazillian category names to english and get rid of the translation df alltogether.
* There are 71 unique category name translations, but product has 73 unique category

```python
products.head()[products['product_category_name'].head().isin(['artes','bebes'])]
```

```python
# Find out the 2 category names which are not present in the category translation df
print(cat_translate['product_category_name'].nunique())
print(products['product_category_name'].nunique())

valid_cat_trans = set(cat_translate['product_category_name'].unique())
all_pdt_cats = set(products['product_category_name'].unique())
invalid_cats = list(all_pdt_cats - valid_cat_trans)
print(invalid_cats)
invalid_cats_pdt = products['product_category_name'].isin(invalid_cats)
previous_rows = products.shape[0]
products = products[~(invalid_cats_pdt)]
print("Previous rows: {}, Current rows: {}, Diff: {}".format(str(previous_rows), str(products.shape[0]), str(previous_rows - products.shape[0])))
```
