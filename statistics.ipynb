# Descriptive Statistics

## Measures of Central Tendency

* Understand data spread around the center in data analysis
* Replace missing values in Data analysis
* Showcase performance of players across their career
* Study salary spread across the company

### Mean

* Average value of a dataset, e.g. average weight, average salary
* Representative of general trend in the sample 
* Gets effected by outliers, the mean tend to move towards the outliers. In this case its not a great representative of the sample.

### Median

* Positional average
* Sort the data, get the center value, in case of even data, get the average of center 2
* Advantage is that median doesn't gets effected by outliers

### Mode

* Most frequently occuring value in a dataset, unimodal
* There can be more than 1 frequently occuring value in a dataset, bimodal, trimodal, multimodal
* It's possible that there is no mode
* e.g. of usage: In a restaurant, you collect the data of how many guests come in a group. Based on that data, you get the mode, then you arrange your tables, mostly they are in groups of 2 or groups of 4


## Measures of Spread

* Gives a representation of how spreaded the data is from lowest to highest point based on the below parameters
* A measure of spread gives us an idea of how well the mean, for example, represents the data. e.g. if the standard deviation is too much, mean is not a good reflection of the sample. but if the standard deviation is too less, then mean is a very good reflection of the sample spread.
* Figure outliers in data
* Standardize data and interpret wrt mean ??
* Monitor control systems for not normal behaviour, too much range/standard deviation tells that something not is not right.

### Range

* Difference of the highest and lowest point
* e.g. if the plant production for plant A has a range of 20, let's say the range for plant B is 4. We can infer that plant B is more consitent. you can rely on plant B, but you can't rely plant A

### Interquartile Range

* Arrange our data in asc form
* Divide the data in 4 equal parts, each part is a quartile. (Q1)25th quartile, (Q2)50th quartile, (Q3)75th quartile & (Q4)100th quartile.
* The middle 50% quartile of the data comes in interquartile range (IQR) i.e. median of (Q3)75th quartile - median of (Q2)50th quartile.
* &******&******* & ******* & **** &
     Q1     Q2        Q3       Q4
     & --   &  --      & --     & Interquartile ranges
* Gives the spread of data in ranges, in the beginning/ middle/end sections of data


### Variance

* Variance is the average of the squared distances from the mean
* Gives an idea about how far the data is spread from the mean

### Standard Deviation

* Tells how close the values in a dataset are to the mean

## Measures of Shape

* Measures of shape describe the distribution(or pattern) of the data within a dataset
* A measure of the extent to which a frequence distribution is concentrated about its mean.
* Measures outliers
* Measures impact of outliers on the overall data

### Symmetric curve

* Bell curve
* Mean = Median = Mode

### Skewness

#### Positively Skewed

* Median < Mean
* Positive Skew because it represents that there are not many outliers on the higher side 

#### Negatively Skewed

* Median > Mean
* Negative Skew because it represents that there are many outliers on the higher side

### Kurtosis

* Demonstrates what is the concentration of data in the center

#### Lepokurtic
#### Mesokurtic
#### Platykurtic

## Contribution of statistics in DS
* Statistics that we apply on sample, descriptive statistics
* Extrapolation of what we learn from descriptive statistics to generalize for entire population is inferential statistics
* Inferential statistics generalizes the larger dataset and applies probability theory to draw a conclusion.

* e.g. 100 insurance agent sell insurance worth `$179,322`, which implies 10,000 insurance agents will also sell `$179,322`. But its not a direct inference. Instead, you say, likelihood of `$179,322` being the mean of the entire population is `78%`
* Inference extrapolation for entire population can not be made with certainity. Its always a probability. With Descriptive statistics we can draw conclusions on a sample with certainity. But, your sample was not the entire population, so you are not sure. We attach a probability number with the conclusion and infer for the entire population.
* This art of finding out a probability number is inferential statistics.
